```python
ages = pd.Series(data['age']).values
b_pressure = pd.Series(data['trestbps']).values
ages_train,ages_test = np.array_split(ages,2)
b_pressure_train,b_pressure_test = np.array_split(b_pressure,2)

ages_train = ages_train.reshape(-1,1)
ages_test = ages_test.reshape(-1,1)

reg =  linear_model.LinearRegression()

reg.fit(ages_train,b_pressure_train)
b_pressure_predicted = reg.predict(ages_test)

data.plot.scatter(x='age',y= 'trestbps')
md.plot(ages_test,b_pressure_predicted,linewidth = 3, color = 'red')
print(reg.coef_)
print(reg.intercept_)
```

In the age vs cholesterol and age vs blood pressure plots, the linear regression model told us that there is a positive correlation in those two sets of relationships suggesting individuals are more succeptible to dangerously high levels or blood pressure and cholesterol as they age. The model for the blood pressure vs cholesterol plot suggests that cholesterol and blood pressure levels are independent of eachother.   

These models suggest that the risk for heart disease increases as a person ages but as seen in the data from step 1, this was not the case. So why do the models say otherwise? I decided to again to an age based approach by figuring out at what age do most blood pressure and cholesterol conditions start.

```python
data['trestbps'].value_counts().head(30).plot.bar(title = "Frequecies of Blood Pressure Levels")
data['trestbps'].describe()
```

```python
data['chol'].value_counts().head(30).plot.bar(title = "Frequecies of Cholesterol Levels")
data['chol'].describe()
```

For visualization purposes I only graphed the first 30 entries. From the two graphs, a blood pressure level of 120 and a cholesterol level of 234 were the most commonly reported. These levels are enough for a patient to be diagnosed with pre-hypertension and high cholesterol. We can use the inverse of the previously constructed linear models to predict at which age is an individual succeptible to reach these levels . 

Starting with the equation y = mx + b we want to solve for x giving us the equation x = (y-b) / m. Using the mean blood pressure / cholesterol level for y, we will use this new equation predict at what age would an individual encounter high levels of blood pressure and cholesterol. The values of b and m are given from our previous linear models. 

```python
#linear model for age vs blood pressure
    #coefficient = 0.48 intercept = 103.67
bp_age = (131 - 103.67) / 0.48

#linear model for age vs cholesterol
    #coefficient = 1.29, intercept = 175.97
chol_age = (246 - 175.97) / 1.29
print(bp_age)
print(chol_age )
print((bp_age + chol_age) / 2)
```

The average age predicted by both models is 55 ~ 56 years old. It can not be coincidence that this result is almost spot on with the 56 year old mean for heart disease from step 1.
